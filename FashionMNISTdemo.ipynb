```python
import keras
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
train_csv = pd.read_csv("/home/suraj/Learning/AdvancedCV_Udemy/fashionmnist/fashion-mnist_train.csv")
```

```python
train_csv.head(5)
```

```python
label = train_csv.loc[:, "label"].values
train_data = train_csv.loc[:, "pixel1":"pixel784"].values
print label.shape, train_data.shape
```

```python
onehot = OneHotEncoder()
label = label.reshape(-1, 1)
print label
label = onehot.fit_transform(label)
```

```python
print label
```

```python
train_data = np.reshape(train_data, (train_data.shape[0], 28, 28, 1))
```

```python
print train_data.shape
```

```python
train_data = train_data.astype(dtype=np.float32)
train_data = train_data/255.0
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Activation, BatchNormalization, Dropout
```

```python
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(3, 3), input_shape=(28, 28, 1)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Conv2D(filters=64, kernel_size=(3, 3)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Conv2D(filters=128, kernel_size=(3, 3)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(10, activation='softmax'))
print model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_data, label, validation_split=0.2, epochs=100, batch_size=32)
```
